# 날짜별로 영화데이터 수집

### 날짜만들기

```python
date = pd.date_range(start = "2021-01-01", end = "2021-05-01")
date
```

```python
#문자형 포매팅
days = date.strftime("%Y%m%d")
days
```

####  진행사항을 볼수있는 함수

```python
from tqdm import tqdm_notebook
```

```python
avg_list = []
movie_title_list = []
rank_list = []
date_list = []
for i in tqdm_notebook(days):
    url1 = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&tg=0&date=" + i
    res = req.get(url1)
    soup = bs(res.text, "lxml")
    movie_title = soup.select("div.tit5 > a") #영화명
    avg = soup.select(".point")   #평점
    
    for j in range(len(movie_title)):
        movie_title_list.append(movie_title[j].text)
        avg_list.append(avg[j].text)
        rank_list.append(j + 1)
        date_list.append(i)
```

```python
print(len(movie_title_list))
print(len(avg_list))
print(len(rank_list))
print(len(date_list))
```

```python
dic = {"순위" : rank_list, "영화명" : movie_title_list, "평점" : avg_list, "날짜" : date_list}
movie_date = pd.DataFrame(dic)
```

```python
movie_date.set_index("순위", inplace=True)
```

```python
movie_date
```

```python
#파일명
movie_date.to_csv("movie_date.csv", encoding="euc-kr")
```
